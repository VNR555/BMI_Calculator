```python
df2 = pd.read_csv(r"C:\Users\Vinod R\Desktop\Tex\text files\Hai 2.csv")
df2
```

**MERGE**

```python
df1.merge(df2) 
#inner join
#extract the matched ones out
```

```python
df1.merge(df2, how = 'inner') 
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID')
```

```python
#both df1 and df2 have a column named FirstName. To avoid confusion and preserve both columns, pandas automatically adds suffixes _x and _y.
```

```python
df1.merge(df2, how = 'outer')
#NaN - Not a Number
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross')
#interview question
```

**JOIN**

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = 'Left', rsuffix = 'Right')
```

```python
#it's better to use indexes while using joins
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = 'Left', rsuffix = 'Right')
df4
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = 'Left', rsuffix = 'Right', how = 'outer')
df4
```

**CONCATENATE**

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join = 'inner')
```

```python
pd.concat([df1,df2], join = 'outer')
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
```
